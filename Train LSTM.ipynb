# Read in dataset and prepare training and testing sets

```python
import pandas as pd

dataframe = pd.read_csv(r'datasets\times&bands.csv')
testing = []
for x in range(1,6):
    testing.append(dataframe[(dataframe.source == x) & (dataframe.destination == 11)][['bandwidth']][-200:])
```

# LSTM predictions
1. load model
2. predict

```python
from tensorflow.keras.models import load_model
from model import TimeSeriesModelCreator
import joblib
```

```python
model_obj = TimeSeriesModelCreator()
```

```python
lstm50 = load_model('models/models/lstm50_1.h5')
scaler50 = joblib.load('models/scalers/scaler50_1.save')
# lstm200 = load_model('models/lstm200model.h5')
# lstm500 = load_model('models/lstm500model.h5')
```

```python
LSTMpredictions = []
for x in range(1,6):
    prediction = model_obj.predict(lstm50,scaler50, testing[x-1])
    LSTMpredictions.append(prediction)
```
